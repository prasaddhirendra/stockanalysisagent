```python
from gnews import GNews
from newspaper import Article

# Initialize GNews
gn = GNews(language='en', country='IN', max_results=5)

# Search News
query = "stock market"
articles = gn.get_news(query)

for article in articles:
    # print("Title:", article['title'])
    # print("Publisher:", article['publisher']['title'])
    # print("Published:", article['published date'])
    # # print("URL:", article['url'])

    try:
        # Fetch full article text using newspaper3k
        news_article = Article(article['url'])
        news_article.download()
        news_article.parse()
        print("Full Article Content:\n", news_article.text[:500], "...")  # first 500 chars
    except Exception as e:
        print("❌ Error fetching article:", e)

    print("="*80)
```
